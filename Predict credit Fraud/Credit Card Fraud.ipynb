```python
# Import all neccesary Libraries and read data
import matplotlib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylot as plt

# Read the train and test data
dataset = pd.read_csv('creditcard.csv')

# Show the first 5 rows
dataset.head(5)
```

```python
count_classes = pd.value_counts(dataset['Class'], sort = True).sort_index()
count_classes.plot(kind = 'bar')
plt.title("Fraud class histogram")
plt.xlabel("Class")
plt.ylabel("Frequency")
```

```python
corrmat = dataset.corr()
f, ax = plt.subplots(figsize=(20, 20))
sns.heatmap(corrmat, vmax=.8, square=True);
```

```python
#class correlation matrix
k = 15 #number of variables for heatmap
cols = corrmat.nlargest(k, 'Class')['Class'].index
f, ax = plt.subplots(figsize=(20, 20))
cm = np.corrcoef(dataset[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
#class correlation matrix
k = 15 #number of variables for heatmap
cols = corrmat.nlargest(k, 'Amount')['Amount'].index
f, ax = plt.subplots(figsize=(20, 20))
cm = np.corrcoef(dataset[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
